```python
def every_30mins():
    print('saving every hour')
    p.run(save='save')
    p.plot_static3(save='save')
    p.plot_func(save='save')
    plt.close('all')
```

```python
def every_minute():
    print('saving every 2 minutes...')
    p.run()
    p.plot_static3()
    p.plot_func()
    plt.close('all')
```

```python
import schedule
import time

schedule.every(2).minutes.do(every_minute)
schedule.every(31).minutes.do(every_30mins)

while True:
    schedule.run_pending()
    time.sleep(1)
```
